Selección y Preparación de Características  
Para predecir el precio de cierre (Close), seleccionaremos las características que creemos que influyen en él, como Open, High, Low y Volume.

```
# Seleccionar las columnas de características y la etiqueta
feature_columns = ["Open", "High", "Low", "Volume"]  # Reemplaza o añade según los atributos disponibles
label_column = "Close"  # Supongamos que queremos predecir el precio de cierre

# Ensamblar las características en un vector
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")

# Escalar las características
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
```

5. División del Conjunto de Datos

```
# Dividir los datos en entrenamiento y prueba (80% - 20%)
train_data, test_data = df_pivot.randomSplit([0.8, 0.2], seed=42)

# Verificar las divisiones
print("Número de filas en entrenamiento:", train_data.count())
print("Número de filas en prueba:", test_data.count())
```

6. Configuración del Pipeline de Spark ML  

Crearemos un pipeline que incluya:  
1. VectorAssembler: Para ensamblar las características en un solo vector.
2. StandardScaler: Para escalar las características.
3. Modelo de Regresión: Podemos empezar con Regresión Lineal y luego probar con otros modelos como Random Forest Regressor.

a. Definimos el Pipeline con Regresión Lineal

```
# Crear un VectorAssembler para ensamblar las características
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")

# Escalar las características
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")

# Definir el modelo de Regresión Lineal
lr = LinearRegression(featuresCol='scaledFeatures', labelCol=label_column)

# Crear el pipeline
pipeline = Pipeline(stages=[assembler, scaler, lr])
```

b. Alternativamente, Definimos el Pipeline con Random Forest Regressor

```
from pyspark.ml.regression import RandomForestRegressor

# Definir el modelo de Random Forest Regressor
rf = RandomForestRegressor(featuresCol='scaledFeatures', labelCol=label_column, numTrees=100)

# Crear el pipeline
pipeline_rf = Pipeline(stages=[assembler, scaler, rf])
```

7. Entrenamiento del Modelo

a. Entrenar con Regresión Lineal

```
# Entrenar el modelo
model = pipeline.fit(train_data)
```

b. Entrenar con Random Forest Regressor

```
# Entrenar el modelo Random Forest
model_rf = pipeline_rf.fit(train_data)
```

8. Hacer Predicciones

a. Predicciones con Regresión Lineal

```
# Hacer predicciones en el conjunto de prueba
predictions = model.transform(test_data)

# Mostrar algunas predicciones
display(predictions.select("Date", "Ticker", label_column, "prediction"))
```

b. Predicciones con Random Forest Regressor

```
# Hacer predicciones en el conjunto de prueba con Random Forest
predictions_rf = model_rf.transform(test_data)

# Mostrar algunas predicciones
display(predictions_rf.select("Date", "Ticker", label_column, "prediction"))
```

9. Evaluación del Modelo
  
Evaluaremos el rendimiento de ambos modelos utilizando métricas como RMSE (Root Mean Squared Error) y R².  
  
  
1. RMSE (Root Mean Squared Error): Mide la diferencia promedio entre los valores predichos y los valores reales, proporcionando una indicación de la precisión del modelo.  
2. R² (Coeficiente de Determinación): Indica la proporción de la varianza en la variable dependiente que es predecible a partir de las variables independientes, reflejando la capacidad explicativa del modelo.

a. Evaluación con Regresión Lineal

```
# Inicializar el evaluador para RMSE
evaluator_rmse = RegressionEvaluator(labelCol=label_column, predictionCol="prediction", metricName="rmse")
rmse = evaluator_rmse.evaluate(predictions)
print(f"Regresión Lineal RMSE: {rmse}")

# Inicializar el evaluador para R²
evaluator_r2 = RegressionEvaluator(labelCol=label_column, predictionCol="prediction", metricName="r2")
r2 = evaluator_r2.evaluate(predictions)
print(f"Regresión Lineal R²: {r2}")
```

El modelo de Regresión Lineal ha mostrado un desempeño excepcional, con un RMSE muy bajo y un R² cercano a 1. Esto sugiere que el modelo es altamente preciso en la predicción del precio de cierre de las acciones y explica casi toda la variabilidad observada en los datos. La simplicidad y la interpretabilidad de la regresión lineal han sido ventajosas en este contexto, proporcionando resultados claros y confiables.

b. Evaluación con Random Forest Regressor

```
# Evaluar RMSE para Random Forest
rmse_rf = evaluator_rmse.evaluate(predictions_rf)
print(f"Random Forest RMSE: {rmse_rf}")

# Evaluar R² para Random Forest
r2_rf = evaluator_r2.evaluate(predictions_rf)
print(f"Random Forest R²: {r2_rf}")
```

Por otro lado, el modelo de Random Forest Regressor también ha demostrado un rendimiento sólido, con un RMSE ligeramente mayor pero aún significativamente bajo, y un R² alto. Aunque este modelo es más complejo y puede capturar relaciones no lineales en los datos. 

##### En este caso específico, la Regresión Lineal ha superado al Random Forest en términos de precisión predictiva. Esto podría deberse a la naturaleza lineal de la relación entre las características seleccionadas y el precio de cierre, haciendo que un modelo lineal sea más adecuado para este conjunto de datos.
